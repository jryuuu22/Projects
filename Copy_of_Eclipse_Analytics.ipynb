```python
SALARY_CAPS = {
    '2022-23': 123655000,
    '2017-18': 99093000
}
```

```python
df_2023 = contract_stats_df.copy()
```

```python
df_meta = player_season_df.copy()
```

```python
#removed whitespace and suffix
df_2023['Player Name'] = df_2023['Player Name'].str.strip()
df_meta['player'] = df_meta['player'].str.strip()

#filtered data for the season of 2023
df_meta_2023 = df_meta[df_meta['season'] == 2023][['player', 'experience', 'age']]

main_df = pd.merge(
    df_2023,
    df_meta_2023,
    left_on='Player Name',
    right_on='player',
    how='left'
)

#If experience is missing, fill with 1 (usually for rookies/unknown)
main_df['experience'] = main_df['experience'].fillna(1)
#ensure its one data for one player (removes duplicates)
main_df = main_df.drop_duplicates(subset=['Player Name'])
```

```python
#Feature

#reset columns
main_df = contract_stats_df.copy()

#merges to get fresh data
main_df = pd.merge(main_df, df_meta_2023, left_on='Player Name', right_on='player', how='left')
main_df['experience'] = main_df['experience'].fillna(1)
main_df = main_df.drop_duplicates(subset=['Player Name'])

#SG-PG changed into SG
main_df['Position'] = main_df['Position'].apply(lambda x: str(x).split('-')[0])

#Salary cap for inflation
cap_2023 = SALARY_CAPS['2022-23']
main_df['Salary_Cap_Percent'] = main_df['Salary'] / cap_2023

#Salary log for better regression performance
main_df['Salary_Log'] = np.log1p(main_df['Salary'])

#If it's a rookie, we flag them if <= 4 years
main_df['Is_Rookie_Scale'] = (main_df['experience'] <= 4).astype(int)

#salary pay scales differ per position
position_dummies = pd.get_dummies(main_df['Position'], prefix='Pos')
main_df = pd.concat([main_df, position_dummies], axis=1)

#keeping relevant columns
columns_to_keep = [
    'Player Name', 'Salary', 'Salary_Cap_Percent', 'Salary_Log',
    'Age', 'experience', 'Is_Rookie_Scale',
    'GP', 'GS', 'MP', 'PTS', 'TRB', 'AST', 'STL', 'BLK',  # Basic Stats
    'FG%', '3P%', 'FT%', 'TS%', 'PER', 'WS', 'BPM', 'VORP' # Advanced Stats
]

#position columns:
pos_cols = [c for c in main_df.columns if c.startswith('Pos_')]
final_target_cols = columns_to_keep + pos_cols

#double checking if columns actually exist
available_cols = [c for c in final_target_cols if c in main_df.columns]
final_df = main_df[available_cols]

print("Data Cleaned Successfully!")
print(f"Total Players: {len(final_df)}")
print(f"Columns: {len(final_df.columns)}")
print(final_df.head())
```
