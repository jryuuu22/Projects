```python
def build_season_df(stats_df, meta_df, season_year, salary_col='Salary', player_name_col='Player Name'):
    df = stats_df.copy()

    # standardize names
    df[player_name_col] = df[player_name_col].astype(str).str.strip()
    meta = meta_df.copy()
    meta['player'] = meta['player'].astype(str).str.strip()

    # season-specific meta
    meta_season = meta[meta['season'] == season_year][['player', 'player_id', 'experience', 'age']]

    # merge meta
    df = df.merge(meta_season, left_on=player_name_col, right_on='player', how='left')

    # ---------- FIX: standardize age column ----------
    # If merge created age_x/age_y, prefer the meta-provided 'age' from player_season_df
    if 'age' not in df.columns:
        # try common alternatives
        for alt in ['Age', 'age_y', 'age_x']:
            if alt in df.columns:
                df['age'] = df[alt]
                break

    # if still missing, create a null column (so code doesn't crash)
    if 'age' not in df.columns:
        df['age'] = np.nan
    # -----------------------------------------------

    # clean experience
    df['experience'] = df['experience'].fillna(1)

    # dedupe
    df = df.drop_duplicates(subset=[player_name_col])

    # position one-hot if present
    if 'Position' in df.columns:
        df['Position'] = df['Position'].apply(lambda x: str(x).split('-')[0])
        pos_dummies = pd.get_dummies(df['Position'], prefix='Pos')
        df = pd.concat([df, pos_dummies], axis=1)

    # salary normalization
    cap = SALARY_CAPS[season_year]
    df['season'] = season_year
    df['Salary_Cap_Percent'] = df[salary_col] / cap
    df['Salary_Log'] = np.log1p(df[salary_col])

    # contract proxy flags
    df['Is_Rookie_Scale'] = (df['experience'] <= 4).astype(int)

    # age curve
    df['Age_Squared'] = df['age'] ** 2

    return df
```

```python
#Feature

#reset columns
main_df = contract_stats_df.copy()

#merges to get fresh data
main_df = pd.merge(main_df, df_meta_2023, left_on='Player Name', right_on='player', how='left')
main_df['experience'] = main_df['experience'].fillna(1)
main_df = main_df.drop_duplicates(subset=['Player Name'])

#SG-PG changed into SG
main_df['Position'] = main_df['Position'].apply(lambda x: str(x).split('-')[0])

#Salary cap for inflation
cap_2023 = SALARY_CAPS[2023]
main_df['Salary_Cap_Percent'] = main_df['Salary'] / cap_2023

#Salary log for better regression performance
main_df['Salary_Log'] = np.log1p(main_df['Salary'])

#If it's a rookie, we flag them if <= 4 years
main_df['Is_Rookie_Scale'] = (main_df['experience'] <= 4).astype(int)

#salary pay scales differ per position
position_dummies = pd.get_dummies(main_df['Position'], prefix='Pos')
main_df = pd.concat([main_df, position_dummies], axis=1)

#keeping relevant columns
columns_to_keep = [
    'Player Name', 'Salary', 'Salary_Cap_Percent', 'Salary_Log',
    'Age', 'experience', 'Is_Rookie_Scale',
    'GP', 'GS', 'MP', 'PTS', 'TRB', 'AST', 'STL', 'BLK',  # Basic Stats
    'FG%', '3P%', 'FT%', 'TS%', 'PER', 'WS', 'BPM', 'VORP' # Advanced Stats
]

#position columns:
pos_cols = [c for c in main_df.columns if c.startswith('Pos_')]
final_target_cols = columns_to_keep + pos_cols

#double checking if columns actually exist
available_cols = [c for c in final_target_cols if c in main_df.columns]
final_df = main_df[available_cols]

print("Data Cleaned Successfully!")
print(f"Total Players: {len(final_df)}")
print(f"Columns: {len(final_df.columns)}")
print(final_df.head())
```
